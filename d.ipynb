```python
def solution(n):
    # n x n 크기의 0으로 초기화된 배열 생성
    matrix = [[0] * n for _ in range(n)]
    # 방향 벡터: 오른쪽, 아래, 왼쪽, 위
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    current_direction = 0  # 현재 방향 (오른쪽)
    row, col = 0, 0  # 시작 위치
    value = 1  # 배열에 채울 값

    while value <= n * n:
        # 현재 위치에 값 채우기
        matrix[row][col] = value
        value += 1

        # 배열 상태 출력
        print(f"채워진 값: {value - 1}, 위치: ({row}, {col}), 방향: {current_direction}")
        for line in matrix:
            print(line)
        print("-" * 20)

        # 다음 위치 계산
        next_row = row + directions[current_direction][0]
        next_col = col + directions[current_direction][1]

        # 다음 위치가 경계를 벗어나거나 이미 채워진 경우 방향 전환
        if (
            next_row < 0 or next_row >= n or
            next_col < 0 or next_col >= n or
            matrix[next_row][next_col] != 0
        ):
            current_direction = (current_direction + 1) % 4  # 다음 방향으로 전환

        # 다음 위치로 이동
        row += directions[current_direction][0]
        col += directions[current_direction][1]

    return matrix
```

```python
def solution(n):
    # n x n 크기의 0으로 초기화된 배열 생성
    matrix = [[0] * n for _ in range(n)]  # n x n 크기의 2차원 리스트를 0으로 초기화
    row, col = 0, 0  # 시작 위치 (첫 번째 행의 첫 번째 열)
    value = 1  # 배열에 채울 시작 값

    # 이동 범위 설정
    top, bottom, left, right = 0, n - 1, 0, n - 1  # 상단, 하단, 좌측, 우측 경계값 초기화

    while value <= n * n:  # 배열에 채울 값이 n^2보다 작거나 같은 동안 반복
        # 오른쪽으로 이동
        for col in range(left, right + 1):  # 현재 행(top)의 좌측부터 우측까지 반복
            matrix[top][col] = value  # 현재 위치에 값을 채움
            value += 1  # 다음 값으로 증가
        top += 1  # 상단 경계를 아래로 한 칸 줄임

        # 아래로 이동
        for row in range(top, bottom + 1):  # 현재 열(right)의 상단부터 하단까지 반복
            matrix[row][right] = value  # 현재 위치에 값을 채움
            value += 1  # 다음 값으로 증가
        right -= 1  # 오른쪽 경계를 왼쪽으로 한 칸 줄임

        # 왼쪽으로 이동
        if top <= bottom:  # 남아 있는 행이 있을 때만 실행
            for col in range(right, left - 1, -1):  # 현재 행(bottom)의 우측부터 좌측까지 반복
                matrix[bottom][col] = value  # 현재 위치에 값을 채움
                value += 1  # 다음 값으로 증가
            bottom -= 1  # 하단 경계를 위로 한 칸 줄임

        # 위로 이동
        if left <= right:  # 남아 있는 열이 있을 때만 실행
            for row in range(bottom, top - 1, -1):  # 현재 열(left)의 하단부터 상단까지 반복
                matrix[row][left] = value  # 현재 위치에 값을 채움
                value += 1  # 다음 값으로 증가
            left += 1  # 왼쪽 경계를 오른쪽으로 한 칸 줄임

    return matrix  # 완성된 나선형 배열 반환

# 테스트
n1 = 4  # 테스트 입력 1
n2 = 5  # 테스트 입력 2

print(solution(n1))  # 결과 출력
print(solution(n2))  # 결과 출력
```

```python
class Cat:
    def __init__(self):
        self.legs = 4
        self.mouth = 1
        self.nose = 1 
        self.eyes = 2
        self.tail = 1

    def 걷기(self):
        print("고양이가 걷고 있습니다")
    
    def 울기(self):
        print("야옹~")
        
    def 먹기(self):
        print("맛있게 먹고 있습니다")
        
    def 정보출력(self):
        print(f"""
        다리 개수: {self.legs}개
        입 개수: {self.mouth}개  
        코 개수: {self.nose}개
        눈 개수: {self.eyes}개
        꼬리 개수: {self.tail}개
        """)

```

```python
class Cat:
    # 클래스 변수 (모든 고양이가 공통으로 가지는 특성)
    legs = 4
    mouth = 1
    nose = 1 
    eyes = 2
    tail = 1

    # 초기화 메서드
    def __init__(self, name, age, skills):
        self.name = name        # 이름
        self.age = age          # 나이 
        self.skills = skills    # 스킬 목록

    # 고양이 정보 출력 메서드
    def cat_info(self):
        print(f"이름: {self.name}")
        print(f"나이: {self.age}살")
        print(f"스킬: {', '.join(self.skills)}")
        print(f"다리: {self.legs}개")
        print(f"입: {self.mouth}개")
        print(f"코: {self.nose}개") 
        print(f"눈: {self.eyes}개")
        print(f"꼬리: {self.tail}개")

# 사용 예시
나비 = Cat("나비", 3, ["점프하기", "낮잠자기", "먹기"])
나비.cat_info()
```
